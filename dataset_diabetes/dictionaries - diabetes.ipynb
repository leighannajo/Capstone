```python
diag_codes = {}
for num in range(390, 460):
    diag_codes[num] = 'circulatory'
    for num in range(460, 520):
        diag_codes[num] = 'respiratory'
        for num in range(520, 580):
            diag_codes[num] = 'digestive'
            for num in range(800, 1000):
                diag_codes[num] = 'injury'
                for num in range(710, 740):
                    diag_codes[num] = 'musculoskeletal'
                    for num in range(580, 630):
                        diag_codes[num] = 'genitourinary'
                        for num in range(140, 240):
                            diag_codes[num] = 'neoplasms'
        
```

```python
diag_codes[239]
```

```python
tester = {785:'circulatory', 786:'respiratory', 787:'digestive',
           788:'genitourinary', 250.00:'diabetes'}
diag_codes.update(tester)
```

```python
diag_codes[240]
```

```python
temp = {}
for num in range(0, 800):
    if num not in diag_codes:
        temp[num] = 'other'
        diag_codes.update(temp) 
```
